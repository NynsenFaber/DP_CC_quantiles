# Set Privacy Budgets

We propose $\varepsilon, \delta$ that are used in SliceQuantiles, and compute the highest $\rho$ for zCDP that implies the same $\varepsilon, \delta$.

```python
from scipy.optimize import root_scalar
import numpy as np


def get_max_rho(eps, delta):
    """
    Get the maximum rho for a given epsilon and delta.
    """
    epsilon_func = lambda rho: rho + np.sqrt(4 * rho * np.log(np.sqrt(np.pi * rho) / delta)) - eps

    # Use a numerical solver to find the root of epsilon_func
    result = root_scalar(epsilon_func, bracket=[1e-10, 10], method='brentq')

    if result.converged:
        return result.root
    else:
        raise ValueError("Failed to find a valid rho for the given epsilon and delta.")


rho = get_max_rho(EPS, DELTA)
print("For eps = {}, delta = {}, max rho = {}".format(EPS, DELTA, rho))
```

# Hyperparameters

```python
RHO = get_max_rho(EPS, DELTA)
M_MAX = 250
M_LIST = range(10, 210, 10)
max_q_list = np.linspace(0, 1, M_MAX + 2)[1:-1]

for dataset in DATASET:
    # add the q_list key
    dataset["q_list"] = max_q_list

# create seeds to generate the quantiles
quantile_seeds = [SEED + i for i in range(NUM_EXPERIMENTS)]
```

```python
records = []  # to store the records of each experiment
```

```python
import math


def get_statistics(X: np.ndarray, quantiles: np.ndarray, estimates: np.ndarray) -> float:
    """
    Compute the maximum rank error and the mean absolute rank error between the true quantiles and the estimated quantiles.
    :param X: data set
    :param quantiles: a list of quantiles (values in [0, 1])
    :param estimates: a list of estimated qth-quantiles (values in X)
    :return: a dictionary with the maximum rank error and the mean absolute rank error
    """
    # sort the data
    X = np.sort(X)
    n = len(X)

    # for each result, get the position of the closest element in the sorted array
    positions = np.searchsorted(X, estimates)
    true_positions = [math.floor(q * n) for q in quantiles]

    # compute the normalized rank error
    rank_errors = np.abs(positions - true_positions)

    # get statistics
    max_rank_error = np.max(rank_errors)
    mean_rank_error = np.mean(rank_errors)
    output = {
        'max_error':  max_rank_error,
        'mean_error': mean_rank_error  # Kaplan et al. uses this
    }
    return output
```

## Run the experiments

```python
from tqdm import tqdm

for swap in (True, False):
    for data_dict in DATASET:
        bound = data_dict["bounds"]
        data = data_dict["data"]
        g = data_dict["g"]
        N = data_dict["n"]
        q_list = data_dict["q_list"]
        for m in tqdm(M_LIST):
            mechanism = SliceQuantile(bound=bound, n=N, m=m, eps=EPS, split=0.5, swap=swap, g=g)
            for i in range(NUM_EXPERIMENTS):
                np.random.seed(quantile_seeds[i])
                Q_LIST = np.sort(np.random.choice(q_list, size=m, replace=False))
                start = time.time()
                estimates = mechanism.approximate_mechanism(data, q_list=Q_LIST, delta=DELTA, verbose=True, sort=False)
                end = time.time()
                statistics = get_statistics(data, quantiles=Q_LIST, estimates=estimates)
                records.append({
                    "m":          m,
                    "max_error":  statistics["max_error"],
                    "mean_error": statistics["mean_error"],
                    "Algorithm":  "SliceQuantiles",
                    "dataset":    data_dict["name"],
                    "swap":       swap,
                    "time":       end - start,
                })
```

# Add density estimation method

```python
def dp_density_estimation(array: np.ndarray,
                          quantiles: list[float],
                          bounds: tuple[float, float],
                          epsilon: float,
                          swap: bool,
                          num_bins: int | str,
                          ) -> np.ndarray:
    """
    Estimate quantiles of the data array using a differentially private histogram
    with Laplace noise added. The histogram is computed over the specified bounds
    with a given number of bins or binning method.

    Parameters:
    - array: input 1D array of numerical data
    - quantiles: list of quantiles to estimate (values between 0 and 1)
    - bounds: tuple specifying the lower and upper bounds for the histogram bins
    - epsilon: privacy parameter controlling the noise magnitude
    - swap: boolean flag indicating sensitivity:
        True -> sensitivity = 2 (e.g., for swap neighboring data changes)
        False -> sensitivity = 1 (standard count sensitivity)
    - num_bins: number of bins (int) or binning strategy (str) for histogram

    Returns:
    - numpy array of quantile estimates corresponding to each quantile in input
    """
    N = len(array)
    # Compute histogram counts over specified bounds and bins
    hist, bins = np.histogram(array, bins=num_bins, range=bounds)

    # Choose sensitivity based on swap definition
    sensitivity = 2.0 if swap else 1.0

    # Add Laplace noise to histogram counts for differential privacy
    noisy_hist = hist + np.random.laplace(loc=0.0, scale=sensitivity / epsilon, size=hist.shape)

    # Compute cumulative distribution function (CDF) from noisy histogram
    cdf = np.cumsum(noisy_hist)

    # Estimate quantiles by finding the first bin where CDF >= quantile * n
    estimates = []
    for q in quantiles:
        target = q * N
        # Default quantile estimate to midpoint of last bin
        quantile_estimate = (bins[-2] + bins[-1]) / 2
        for i in range(len(cdf)):
            if cdf[i] >= target:
                # Update quantile estimate to midpoint of corresponding bin
                quantile_estimate = (bins[i] + bins[i + 1]) / 2
                break
        estimates.append(quantile_estimate)

    return np.array(estimates)
```

```python
from tqdm import tqdm

for swap in (True, False):
    for data_dict in DATASET:
        bound = data_dict["bounds"]
        data = data_dict["data"]
        g = data_dict["g"]
        N = data_dict["n"]
        q_list = data_dict["q_list"]

        num_bins_list = [200] # first try with 200 bins
        num_bins_list.append("auto")
        names =  ["Density Estimation 200 bins", "Density Estimation auto bins"]

        for m in tqdm(M_LIST):
            for num_bins, name in zip(num_bins_list, names):
                for i in range(NUM_EXPERIMENTS):
                    np.random.seed(quantile_seeds[i])
                    Q_LIST = np.sort(np.random.choice(q_list, size=m, replace=False))
                    start = time.time()
                    estimates = dp_density_estimation(array=data, quantiles=Q_LIST, bounds=bound,
                                                      epsilon=EPS, swap=swap, num_bins=num_bins)
                    end = time.time()
                    statistics = get_statistics(data, quantiles=Q_LIST, estimates=estimates)
                    records.append({
                        "m":          m,
                        "max_error":  statistics["max_error"],
                        "mean_error": statistics["mean_error"],
                        "Algorithm":  name,
                        "dataset":    data_dict["name"],
                        "swap":       swap,
                        "time":       end - start,
                    })
```
